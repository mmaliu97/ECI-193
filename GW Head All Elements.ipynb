```python
#init data

df_init = pd.read_csv("FM2Sim_GW_HeadAll_Elements_initial.csv", 
                 parse_dates=True, 
                 index_col="TIME")
df_init = df_init.drop(columns='LAYER')
df0U_init = df_init.iloc[::3, :]
df1C_init = df_init.iloc[1::3, :]
df2A_init = df_init.iloc[2::3, :]

df_int = pd.read_csv("FM2Sim_GW_HeadAll_Elements_intermediate.csv", 
                 parse_dates=True, 
                 index_col="TIME")
df_int = df_int.drop(columns='LAYER')
df0U_int = df_int.iloc[::3, :]
df1C_int = df_int.iloc[1::3, :]
df2A_int = df_int.iloc[2::3, :]

df_rob = pd.read_csv("FM2Sim_GW_HeadAll_Elements_robust.csv", 
                 parse_dates=True, 
                 index_col="TIME")
df_rob = df_rob.drop(columns='LAYER')
df0U_rob = df_rob.iloc[::3, :]
df1C_rob = df_rob.iloc[1::3, :]
df2A_rob = df_rob.iloc[2::3, :]


df_base = pd.read_csv("FM2Sim_GW_HeadAll_Elements_base.csv", 
                 parse_dates=True, 
                 index_col="TIME")
df_base = df_base.drop(columns='LAYER')
df0U_base = df_base.iloc[::3, :]
df1C_base = df_base.iloc[1::3, :]
df2A_base = df_base.iloc[2::3, :]
```

```python
element = '3512'
fig_points=plt.figure(figsize=(12, 7), dpi=160)


base = df2A_base[element]
plt.plot((df2A_init[element]-base)/base, label = "Initial")
plt.plot((df2A_int[element]-base)/base, label = "Intermediate")
plt.plot((df2A_rob[element]-base)/base, label = "Robust")


plt.xlabel("Year",fontsize = 12)
plt.ylabel("Change in groundwater level depth (relative to baseline)",fontsize = 12)
plt.title("Groundwater levels compared to baseline for Element 3512 for layer 2A",fontsize = 18)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(fontsize=12)
plt.show()

```

```python
element = '3512'
fig_points=plt.figure(figsize=(12, 7), dpi=160)


plt.plot(df0U_base.mean(axis=1), label = "0U")
plt.plot(df1C_base.mean(axis=1), label = "1C")
plt.plot(df2A_base.mean(axis=1), label = "2A")


plt.xlabel("Year",fontsize = 12)
plt.ylabel("Groudnwater level depth (relative to sea level)",fontsize = 12)
plt.title("Average groundwater level using Baseline for sublayer conditions",fontsize = 18)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(fontsize=12)
plt.show()

```
