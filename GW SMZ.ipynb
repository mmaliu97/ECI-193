```python
# functions

# making function to get average
def avg_appender(data):
    avg = []

    for i in range(len(data)):
        w=data.iloc[i].mean(axis=0)
        avg.append(w)
    
    data['average'] = avg
    
    return data
## YOU CAN MAKE MORE FUNCTIONS THAT APPLY TO THE WHOLE TIME SERIES
```

## Tabulating number of positive and negative differences

```python
def difference_maker(data):
    diff = []
    for i in range(len(data['average'])-1):
        difference = data['average'][i+1] - data['average'][i]
        diff.append(difference)
    diff.append(np.nan)
    return diff
```

```python
#DAC
DAC_SMZ_base_data['difference'] = difference_maker(DAC_SMZ_base_data)
DAC_SMZ_init_data['difference_wrt_base'] = difference_maker(DAC_SMZ_init_data) - DAC_SMZ_base_data['difference']
DAC_SMZ_int_data['difference_wrt_base'] = difference_maker(DAC_SMZ_int_data) - DAC_SMZ_base_data['difference']
DAC_SMZ_rob_data['difference_wrt_base'] = difference_maker(DAC_SMZ_rob_data) - DAC_SMZ_base_data['difference']

#GWD
GWD_SMZ_base_data['difference'] = difference_maker(GWD_SMZ_base_data)
GWD_SMZ_init_data['difference_wrt_base'] = difference_maker(GWD_SMZ_init_data) - GWD_SMZ_base_data['difference']
GWD_SMZ_int_data['difference_wrt_base'] = difference_maker(GWD_SMZ_int_data) - GWD_SMZ_base_data['difference']
GWD_SMZ_rob_data['difference_wrt_base'] = difference_maker(GWD_SMZ_rob_data) - GWD_SMZ_base_data['difference']

#SUBS
SUBS_SMZ_base_data['difference'] = difference_maker(SUBS_SMZ_base_data)
SUBS_SMZ_init_data['difference_wrt_base'] = difference_maker(SUBS_SMZ_init_data) - SUBS_SMZ_base_data['difference']
SUBS_SMZ_int_data['difference_wrt_base'] = difference_maker(SUBS_SMZ_int_data) - SUBS_SMZ_base_data['difference']
SUBS_SMZ_rob_data['difference_wrt_base'] = difference_maker(SUBS_SMZ_rob_data) - SUBS_SMZ_base_data['difference']

```

```python
DAC_SMZ_init_data_no_ele = DAC_SMZ_init_data[['average','difference_wrt_base']].resample('A-SEP',origin = 'start_day').sum()
DAC_SMZ_int_data_no_ele = DAC_SMZ_int_data[['average','difference_wrt_base']].resample('A-SEP',origin = 'start_day').sum()
DAC_SMZ_rob_data_no_ele = DAC_SMZ_rob_data[['average','difference_wrt_base']].resample('A-SEP',origin = 'start_day').sum()

GWD_SMZ_init_data_no_ele = GWD_SMZ_init_data[['average','difference_wrt_base']].resample('A-SEP',origin = 'start_day').sum()
GWD_SMZ_int_data_no_ele = GWD_SMZ_int_data[['average','difference_wrt_base']].resample('A-SEP',origin = 'start_day').sum()
GWD_SMZ_rob_data_no_ele = GWD_SMZ_rob_data[['average','difference_wrt_base']].resample('A-SEP',origin = 'start_day').sum()

SUBS_SMZ_init_data_no_ele = SUBS_SMZ_init_data[['average','difference_wrt_base']].resample('A-SEP',origin = 'start_day').sum()
SUBS_SMZ_int_data_no_ele = SUBS_SMZ_int_data[['average','difference_wrt_base']].resample('A-SEP',origin = 'start_day').sum()
SUBS_SMZ_rob_data_no_ele = SUBS_SMZ_rob_data[['average','difference_wrt_base']].resample('A-SEP',origin = 'start_day').sum()
```

### Binning each category

```python
def binner(df,intensity):
    df["diff_categories"] = pd.cut(df.difference_wrt_base, 
                                        bins = [df['difference_wrt_base'].min(),
                                         0-abs(df['difference_wrt_base'].min()*0.25),
                                         df['difference_wrt_base'].max()*0.25,       
                                         df['difference_wrt_base'].max()], 
                                        labels = ['Negative Change in GW ' + intensity,
                                                  '10% Change in GW ' + intensity,
                                                  'Positive Change in GW ' + intensity])
```

```python
binner(DAC_SMZ_init_data_no_ele, 'initial')
binner(DAC_SMZ_int_data_no_ele, 'intermediate')
binner(DAC_SMZ_rob_data_no_ele, 'robust')
```

```python
init_vals = DAC_SMZ_init_data_no_ele.groupby('diff_categories').count().reset_index()
int_vals = DAC_SMZ_int_data_no_ele.groupby('diff_categories').count().reset_index()
rob_vals = DAC_SMZ_rob_data_no_ele.groupby('diff_categories').count().reset_index()
```

## Plotting 95th percentile

```python
def ninefive_percent(data):
    ls = []
    for i in range(len(data)):
        arr = data.iloc[i][:-2]
        perc = np.percentile(arr, 95)
        ls.append(perc)
        
    data['95_perc'] = ls
    

```

```python
# Getting 95th percentile
ninefive_percent(DAC_SMZ_init_data)
ninefive_percent(DAC_SMZ_base_data)
ninefive_percent(DAC_SMZ_rob_data)
ninefive_percent(DAC_SMZ_int_data)

```

```python
figure(figsize=(8, 6), dpi=80)
plt.plot(DAC_SMZ_base_data['95_perc'], label = "Base")
plt.plot(DAC_SMZ_init_data['95_perc'], label = "Initial")
plt.plot(DAC_SMZ_int_data['95_perc'], label = "Intermediate")
plt.plot(DAC_SMZ_rob_data['95_perc'], label = "Robust")
# plt.axhline(y=0.4, color='k', linewidth=1)

plt.xlabel("Year")
plt.ylabel("Groundwater Depth")
plt.title("95th Percentile of GW Depth per Scenario")
plt.legend()
plt.show()
```
