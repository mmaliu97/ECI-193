```python
import numpy as np
import csv
import pandas as pd
from shapely.geometry import Polygon
import cv2
from cv2 import contourArea
import geopandas as gpd
```

```python
data= pd.read_csv("Nodes_Elements.csv")
data = data.astype({"Element ID": int})
strat = pd.read_csv("FM2Sim_Stratigraphy.csv")
data['elevation'] = strat['Ground Surface Elevation']
```

```python
node_coord = data[["Node ID", "X","Y","elevation"]]

node_dict = node_coord.set_index('Node ID').T.to_dict('list')

node_dict[0]=0,0
```

```python
def avg(lst):
    return sum(lst) / len(lst)
```

```python
# Dictionary starts from index of 1

data_node = node_dict
data_node
```

```python
node_coord['elevation'][50:123]
```

```python
avg_xcoord = []
avg_ycoord = []
area = []
elev = []
for i in range(len(data)):
    
    n1=data['Node 1'][i]
    n2=data['Node 2'][i]
    n3=data['Node 3'][i]
    n4=data['Node 4'][i]
    if n4 != 0:
        
        c1 = node_dict[n1]
        c2 = node_dict[n2]
        c3 = node_dict[n3]
        c4 = node_dict[n4]
        
        x_list = [c1[0],c2[0],c3[0],c4[0]]
        y_list = [c1[1],c2[1],c3[1],c4[1]]
        elevation = [c1[2],c2[2],c3[2],c4[2]]

        # getting polygon data
        pgon = Polygon(zip(x_list, y_list)) # Assuming the OP's x,y coordinates

        d = [num for num in x_list if num]
        e = [num for num in y_list if num]
        f = [num for num in elevation if num]

        avg_x = avg(d)
        avg_y = avg(e)
        avg_ele = avg(f)
        print(avg_ele)


        area.append(pgon.area)
        avg_xcoord.append(avg_x)
        avg_ycoord.append(avg_y)
        elev.append(avg_ele)
    
    else:
        c1 = node_dict[n1]
        c2 = node_dict[n2]
        c3 = node_dict[n3]
        
        x_list = [c1[0],c2[0],c3[0]]
        y_list = [c1[1],c2[1],c3[1]]
        elevation = [c1[2],c2[2],c3[2]]

        # getting polygon data
        pgon = Polygon(zip(x_list, y_list)) # Assuming the OP's x,y coordinates

        d = [num for num in x_list if num]
        e = [num for num in y_list if num]
        f = [num for num in elevation if num]

        avg_x = avg(d)
        avg_y = avg(e)
        avg_ele = avg(f)
        print(avg_ele)

        area.append(pgon.area)
        avg_xcoord.append(avg_x)
        avg_ycoord.append(avg_y)
        elev.append(avg_ele)

    
    
```

```python
elev
```

```python
data['avg_xcoord'] = avg_xcoord
data['avg_ycoord'] = avg_ycoord
data['ele_area'] = area
data['elevation'] = elev
```
